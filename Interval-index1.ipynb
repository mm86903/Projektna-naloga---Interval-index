```python
import numpy as np
from itertools import combinations
from tqdm import tqdm
import matplotlib.pyplot as plt
import math
```

```python
def interval_set(G, u, v):
    d = dict(nx.single_source_shortest_path_length(G, u))
    d_v = dict(nx.single_source_shortest_path_length(G, v))
    d_uv = d_v.get(u, math.inf)
    I = [w for w in G.nodes() if d.get(w, math.inf) + d_v.get(w, math.inf) == d_uv]
    return I
```

```python
G = nx.path_graph(5)  # vozlišča 0,1,2,3,4
```

```python
I = interval_set(G, 0, 4)
print("Interval med 0 in 4:", I)
```
